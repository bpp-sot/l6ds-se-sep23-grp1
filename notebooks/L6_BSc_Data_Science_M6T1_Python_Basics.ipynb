## **6. Lists**

Data structures are a way of organising and storing data. Python offers various data structures, both primitive and non-primitive.

Imagine a list as a backpack. You can put various items like books, pens, or even another smaller bag inside this backpack. Just like how you can access each item in your backpack by remembering its position, you can access items in a Python list using an index. And you're free to add or remove items whenever you want, making it flexible.
image.png

Let's create a list and print out the first element in our list.

```python
# Creating a list
my_backpack = ["book", "pen", "calculator"]

# Accessing the first element
first_item = my_backpack[0]  # Output: 'book'
print(first_item)
```

We can add and remove items from our list too. Using the `append()` and `insert()` functions.

```python
# Our backpack
my_backpack = ["book", "pen", "calculator"]

# Adding an element using the append, this adds to the end of the list
my_backpack.append('calculator')
print(my_backpack) # showing our list after adding a 'calculator'

# We can use insert to put things in our list in specific positions
my_backpack.insert(1, 'water bottle')
print(my_backpack)
```

â“ðŸ¤” You can see that 'water bottle' was inserted into the position 1 (second index in our list).

We can also remove items from our list with the `remove()` and `pop()` functions.

```python
# Our backpack
my_backpack = ['book', 'water bottle', 'pen', 'calculator', 'calculator']

# Let's remove the item 'book' and print the contents of our list.
my_backpack.remove('book')
print(my_backpack)

# Let's use pop() to remove the item in index 2 (third position) in our list
my_backpack.pop(2)
print(my_backpack)
```

## **7. Dictionaries**

Dictionaries are collections of key-value pairs that are unordered and mutable.

Think of a dictionary as a labelled filing cabinet. Each file has a unique label or "key," and inside each file, you have specific "values" or documents. You don't care about the order of the files; you just need to know the label to quickly grab the corresponding documents. In Python, dictionaries allow you to store and access data in key-value pairs, offering quick retrieval.
image.png

```python
# Create our dictionary
my_filing_cabinet = {'passport': 'important', 'snacks': 'yummy', 'keys': 'essential'}
print(my_filing_cabinet)
```

â“ðŸ¤” Which are the keys and values here?

#### **Accessing Items**

```python
# Accessing elements in our dictionary
passport_details = my_filing_cabinet['passport']
print(passport_details)
```

â“ðŸ¤” Notice how we use our key, '`passport`' to get the value associated with it, which is 'important'.

#### **Adding and Removing Items**

```python
# Let's add a new key and value
my_filing_cabinet['money'] = 'needed'
print(my_filing_cabinet)
```

```python
# Now let's delete the snacks key, this deletes both key and value
del my_filing_cabinet['snacks']
print(my_filing_cabinet)
```

## **8. Tuples**

Tuples are ordered, immutable collections that can contain items of different types.

A tuple is like a treasure chest with compartments. Once you put your valuables (items) in it and lock it, you can't change what's inside. However, you can look into each compartment to see what you stored. In Python, a tuple is an ordered collection like a list, but it's immutable, meaning you can't alter its contents once created.

image.png

#### **Creating a Tuple**

```python
# Our tuple
my_treasure_chest = ('gold', 'jewels', 'ancient scroll')
print(my_treasure_chest)
```

â“ðŸ¤” Notice it's encloded by regular parenthesis or brackets `()`, whereas lists used square brackets `[]` and dictionaries used curly brackets `{}`

```python
first_treasure = my_treasure_chest[0]
print(first_treasure)
```

 â“ðŸ¤” Notice we haven't shown you how to add and remove value from a Tuple, that's because you can't add or remove items once a tuple is created.
